## Functions

```python
def clean_tweet(tweet):
    '''
    Utility function to clean tweet text by removing links and special characters
    (except punctuation, apostrophes, and monetary symbols) using simple regex statements.
    '''
    return ' '.join(re.sub(r"(@[A-Za-z0-9]+)|([^0-9A-Za-z \t\U0001F600-\U0001F64F.,!?':;‚Äô$‚Ç¨¬£])|(\w+:\/\/\S+)", " ", tweet).split())
```

```python
def clean_text(text):
    # Remover stop words
    text = ' '.join([word for word in text.split() if word.lower() not in stop_words])
    return text
```

```python
def classify_sentiment_textblob(text):
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity
    if polarity > 0:
        return 'positive'
    elif polarity < 0:
        return 'negative'
    else:
        return 'neutral'
```

```python
def classify_sentiment_vader(polarity_scores):
    '''
    Classify sentiment based on the compound score from vaderSentiment polarity scores.
    '''
    compound = polarity_scores['compound']
    if compound >= 0.05:
        return 'positive'
    elif compound <= -0.05:
        return 'negative'
    else:
        return 'neutral'
```

#### using textblob

```python
text_1 = "The movie was so awesome."
text_2 = "The food here tastes terrible."

#Determining the Polarity 
p_1 = TextBlob(text_1).sentiment.polarity
p_2 = TextBlob(text_2).sentiment.polarity

#Determining the Subjectivity
s_1 = TextBlob(text_1).sentiment.subjectivity
s_2 = TextBlob(text_2).sentiment.subjectivity

print("Polarity of Text 1 is", p_1)
print("Polarity of Text 2 is", p_2)
print("Subjectivity of Text 1 is", s_1)
print("Subjectivity of Text 2 is", s_2)
```

```python
# Frase a ser analisada
text = "How do you express gratitude for one‚Äôs safety when the perils posed by the American police state grow more treacherous by the day?"

# Calcular a polaridade usando TextBlob
blob = TextBlob(text)
polarity = blob.sentiment.polarity

# Classificar o sentimento da frase
sentiment = classify_sentiment_textblob(polarity)

print(f"Text: {text}")
print(f"Polarity: {polarity}")
print(f"Sentiment: {sentiment}")
```

```python
df_clean['sentiment_textblob'] = df_clean['text'].apply(lambda x: classify_sentiment_textblob(x))
```

```python
sentiment = SentimentIntensityAnalyzer()
text_1 = "The book was a perfect balance between wrtiting style and plot."
text_2 =  "The pizza tastes terrible."
sent_1 = sentiment.polarity_scores(text_1)
sent_2 = sentiment.polarity_scores(text_2)
print("Sentiment of text 1:", sent_1)
print("Sentiment of text 2:", sent_2)
```

```python
# Inicializar o analisador de sentimento
analyzer = SentimentIntensityAnalyzer()

# Frase a ser analisada
text = "How do you express gratitude for one‚Äôs safety when the perils posed by the American police state grow more treacherous by the day?"

# Calcular as pontua√ß√µes de polaridade
polarity_scores = analyzer.polarity_scores(text)

# Fun√ß√£o para classificar o sentimento
def classify_sentiment(polarity_scores):
    compound = polarity_scores['compound']
    if compound >= 0.05:
        return 'positive'
    elif compound <= -0.05:
        return 'negative'
    else:
        return 'neutral'

# Classificar o sentimento da frase
sentiment = classify_sentiment(polarity_scores)

print(f"Text: {text}")
print(f"Polarity Scores: {polarity_scores}")
print(f"Sentiment: {sentiment}")
```
